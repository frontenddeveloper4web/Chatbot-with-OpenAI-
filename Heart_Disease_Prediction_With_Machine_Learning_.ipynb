```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X=data.drop(columns='target', axis=1)
y=data['target']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2, stratify=y)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
x_train_prediction = model.predict(X_train)
training_data_acc = accuracy_score(x_train_prediction, y_train)
```

```python
# Training Data Accuracy
training_data_acc
```

```python
x_test_prediction = model.predict(X_test)
test_data_acc = accuracy_score(x_test_prediction, y_test)
```

```python
test_data_acc
```

```python
# Predictive Model
input_data = (37,1,2,130,250,0,1,187,0,3,5,0,0)

df_array = np.asarray(input_data)
reshaped_data = df_array.reshape(1,-1)

prediction = model.predict(reshaped_data)
```
